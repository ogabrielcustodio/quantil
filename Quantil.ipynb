# Quantil

É uma medida estatística usada para dividir um conjunto de dados em 4 partes iguais, sendo importante para o estudo e entendimento do comportamento dos dados, ajudando a entender a variação e a tendência central dos dados.

image.png

## Características importantes do quantil

- São dividos em 4 partes iguais, conhecidas como **quartis**
- O 1º quartil corresponde a 25% dos dados
- O 2º quartil corresponde a 50% dos dados (mediana)
- O 3º quartil corresponde a 75% dos dados
- Os dados que estão no 1º quartil são inferiores ao 2º quartil
- Os dados que estão no 3º quartil são superiores ao 2º quartil

## Entendendo o quantil pela matemática

Para calcular um quantil é preciso ordenar os dados para que seja retirado o 2º quartil. O 2º quartil é também a mediana, logo, se estamos em um conjunto de dados com quantidade ímpar, a mediana é o valor central e único, contudo, se estamos em um conjunto de dados com quantidade par, a mediana será o resultado da média entre os dois valores do meio.

Logo em seguida retiramos o 1º quartil e o 3º quartil que são a mediana dos dois conjuntos os quais foram definidos pelo 2º quartil.

## Entendendo o quantil pelo Python

No Python calculamos o Quantil usando a biblioteca Pandas ou Numpy, ambas tem métodos que calculam o quantil.

## Método Quantile - Pandas

O pandas calcula o quantil utilizando a interpolação linear nos valores ordenados da série de dados.
Interpolação linear é estimar um valor entre dois valores conhecidos ou entre dois pontos de uma reta. O uso da interpolação linear ocorre quando a posição do quantil não é um número inteiro, ou seja, é um número fracionário (decimal).

**Etapas do cálculo do quantil no pandas**
- Ordena os dados
- Calcula a posição do quantil: p = q X (n-1), se o resultado for um número inteiro, o quantil é o valor correspondente aquela posição na série de dados, senão é feita a interpolação linear
- Interpolação linear: Qp = (1-frac) x Vp + frac x Vp



```python
import pandas as pd
```

```python
dados = [1,2,3,4,5]
series = pd.Series(dados)
```

## 1º Quartil

```python
series.quantile(0.25)
```

## 2º Quartil

```python
series.quantile(0.5)
```

## 3º Quartil

```python
series.quantile(0.75)
```

## Método Quantile - Numpy

O numpy calcula o quantil de forma similar ao pandas, contudo, ocorre a possibilidade de especificar o método de interpolação, podendo ser:

- linear: um valor entre dois pontos conhecidos
- lower: um valor mais baixo entre dois pontos conhecidos
- higher: um valor mais alto entre dois pontos conhecidos
- nearest: um valor mais próximo entre dois pontos conhecidos
- midpoint: a média dos valores lower e higher

```python
import numpy as np
```

```python
dados = [1,2,3,4,5]
```

## Cálculo do quantis no numpy usando métodos de interpolação diferentes

```python
np.quantile(dados, 0.25, interpolation='linear')
```

```python
np.quantile(dados, 0.50, interpolation='lower')
```

```python
np.quantile(dados, 0.25, interpolation='higher')
```

```python
np.quantile(dados, 0.75, interpolation='nearest')
```

```python
np.quantile(dados, 0.75, interpolation='midpoint')
```
